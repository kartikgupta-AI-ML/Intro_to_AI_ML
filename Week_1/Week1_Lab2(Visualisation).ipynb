```python
import pandas as pd
```

```python
df = pd.read_csv('ipl_dataset.csv')
```

```python
import matplotlib.pyplot as plt
```

# Who are the top run-scorers?

```python
plt.figure(figsize=(30,10))
plt.bar(df['Player'], df['Runs'])
```

# Is there any relationship between a batsman's strike rate and runs?

```python
plt.scatter(df['Strike_Rate'], df['Runs'], c=df['Matches'], s=100)  # Color by matches, size bigger for more games
plt.xlabel('Strike Rate')
plt.ylabel('Runs')
plt.colorbar(label='Matches Played')  # Explain colors
# plt.show()

for i in range(len(df)):
    plt.text(
        df['Strike_Rate'][i],
        df['Runs'][i],
        df['Player'][i],
        fontsize=8
    )
```
