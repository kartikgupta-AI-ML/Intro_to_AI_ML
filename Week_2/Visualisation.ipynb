```python
import pandas as pd
import matplotlib.pyplot as plt
ipl_df = pd.read_csv('/content/ipl_dataset.csv')
```

```python
ipl_df
```

# Who are the top run-scorers?

```python
top5 = ipl_df.sort_values('Runs', ascending=False).head(5)


# Step 3: Add colors and rotation (incremental: make it pretty and fix overlap)
plt.bar(top5['Player'], top5['Runs'], color='blue')  # Color bars blue, like team jerseys
plt.title('Top 5 IPL Run-Scorers')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.xticks(rotation=45)  # Rotate names if they overlap
plt.show()
```

# Is there any relationship between a batsman's strike rate and runs?

```python
plt.scatter(ipl_df['Strike_Rate'], ipl_df['Runs'])  # Dots: strike on x, runs on y
# plt.show()

# Step 2: Add labels and title
plt.scatter(ipl_df['Strike_Rate'], ipl_df['Runs'])
plt.title('Strike Rate vs Runs Scored')
plt.xlabel('Strike Rate')
plt.ylabel('Runs')
# plt.show()

# Step 3: Add colors by team and size by matches (incremental: add context)
plt.scatter(ipl_df['Strike_Rate'], ipl_df['Runs'], c=ipl_df['Matches'], cmap='viridis', s=ipl_df['Matches']*10)  # Color by matches, size bigger for more games
plt.title('Strike Rate vs Runs Scored')
plt.xlabel('Strike Rate')
plt.ylabel('Runs')
plt.colorbar(label='Matches Played')  # Explain colors
# plt.show()

for i in range(len(ipl_df)):
    plt.text(
        ipl_df['Strike_Rate'][i],
        ipl_df['Runs'][i],
        ipl_df['Player'][i],
        fontsize=8
    )
```

# How consistent is Virat's Performance across matches?

```python
# Synthetic varied runs per match (highs/lows, sums to total runs)
per_match_runs = [100, 20, 90, 110, 30, 60, 80, 100, 120, 60, 36, 36, 47, 52]  # Varied for trend
match_no = range(1,15)

# Step 3: Add markers, color, and grid (incremental: highlight trend)
plt.plot(match_no, per_match_runs)  # Line: matches implied on x, cumulative runs on y
plt.show()

# Step 2: Add labels and title
plt.plot(match_no, per_match_runs)
plt.title('Cumulative Runs for Virat Kohli')
plt.xlabel('Match Number')
plt.ylabel('Cumulative Runs')
plt.show()

# Step 3: Add markers and grid (incremental: make trends clear)
plt.plot(match_no, per_match_runs, marker='o', linestyle='--', color='green')  # Dots on line, dashed, green
plt.title('Cumulative Runs for Virat Kohli')
plt.xlabel('Match Number')
plt.ylabel('Cumulative Runs')
plt.grid(True)  # Grid for easier reading
plt.show()
```

# How is the range of averages of players?

```python
# Step 1: Simple histogram
plt.hist(ipl_df['Average'])  # Bins of averages
# plt.show()

# Step 2: Add title, labels, and bins
plt.hist(ipl_df['Average'], bins=5)  # 5 buckets
plt.title('Distribution of Batting Averages')
plt.xlabel('Average')
plt.ylabel('Number of Players')
# plt.show()

# Step 3: Add color and edge (incremental: highlight)
plt.hist(ipl_df['Average'], bins=5, color='orange', edgecolor='black')  # Orange bars, black borders
plt.title('Distribution of Batting Averages')
plt.xlabel('Average')
plt.ylabel('Number of Players')
# plt.show()
```

```python
import random

virat_run = random.sample(range(10, 101), 14)
rohit_run = random.sample(range(10, 101), 14)

matches = range(1,15)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # No shared y (Runs vs Fifties differ)
axs[0].plot(matches, virat_run, color='skyblue', marker='o', linestyle='--',)
axs[0].set_title('Runs by Top 5 Players')
axs[0].set_xlabel('Player')
axs[0].set_ylabel('Runs')# Consistent rotation and size
axs[0].grid(True, axis='y', linestyle='--', alpha=0.7) # Grid for readability

axs[1].plot(matches, rohit_run, color='lightgreen', marker='o', linestyle='--',)
axs[1].set_title('Fifties by Top 5 Players')
axs[1].set_xlabel('Player')
axs[1].set_ylabel('Fifties')  # Consistent rotation and size
axs[1].grid(True, axis='y', linestyle='--', alpha=0.7)
```

# Compare runs and fifties of top 5 players

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming ipl_df is loaded (from your latest dataset)
# Select top 5 players by runs to reduce clutter
top5 = ipl_df.sort_values('Runs', ascending=False).head(5)

# Step 1: Simple subplots (bare bones, top 5 only)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # Wider figure to space out labels
axs[0].bar(top5['Player'], top5['Runs'])  # Left: Runs
axs[1].bar(top5['Player'], top5['Fifties'])  # Right: Fifties
# plt.show()

# Step 2: Add titles, labels, and rotation (incremental: readable)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # Keep wider figure
axs[0].bar(top5['Player'], top5['Runs'])
axs[0].set_title('Runs by Top 5 Players')
axs[0].set_xlabel('Player')
axs[0].set_ylabel('Runs')
axs[0].tick_params(axis='x', rotation=45, labelsize=8) # Corrected

axs[1].bar(top5['Player'], top5['Fifties'])
axs[1].set_title('Fifties by Top 5 Players')
axs[1].set_xlabel('Player')
axs[1].set_ylabel('Fifties')
axs[1].tick_params(axis='x', rotation=45, labelsize=8)  # Corrected line # Smaller font, 45 rotation
plt.tight_layout()  # Adjust spacing
plt.show()

# # Step 3: Add colors and shared y-scale (incremental: polish)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # No shared y (Runs vs Fifties differ)
axs[0].bar(top5['Player'], top5['Runs'], color='skyblue')
axs[0].set_title('Runs by Top 5 Players')
axs[0].set_xlabel('Player')
axs[0].set_ylabel('Runs')
axs[0].tick_params(axis='x', rotation=45, labelsize=8)  # Consistent rotation and size
axs[0].grid(True, axis='y', linestyle='--', alpha=0.7) # Grid for readability

axs[1].bar(top5['Player'], top5['Fifties'], color='lightgreen')
axs[1].set_title('Fifties by Top 5 Players')
axs[1].set_xlabel('Player')
axs[1].set_ylabel('Fifties')
axs[1].tick_params(axis='x', rotation=45, labelsize=8)  # Consistent rotation and size
axs[1].grid(True, axis='y', linestyle='--', alpha=0.7)  # Grid for readability
# plt.tight_layout()
# plt.show()
```

# How consistent are the players of each team

```python
import seaborn as sns

# Step 2: Add title and rotation
sns.boxplot(x='Team', y='Runs', data=ipl_df)
plt.title('Runs Distribution by Team')

# Step 3: Add color palette (incremental: theme it)
sns.boxplot(x='Team', y='Runs', data=ipl_df, palette='Set2')  # Colors for teams
plt.title('Runs Distribution by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Prep: Correlation matrix (like finding links)
corr = ipl_df[['Runs', 'Average', 'Strike_Rate', 'Highest_Score', 'Centuries', 'Fifties']].corr()

# Step 1: Simple heatmap
sns.heatmap(corr)  # Colors for correlations
plt.show()

# Step 2: Add annotations and title
sns.heatmap(corr, annot=True)  # Numbers on cells
plt.title('Correlation Heatmap of IPL Stats')
plt.show()
```
